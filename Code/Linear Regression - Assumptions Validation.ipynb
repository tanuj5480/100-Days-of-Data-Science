```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

%matplotlib inline
```

### Linear Regression Assumptions

* Linear relation between features and the dependant variable
* No Outliers
* No high-leverage points
* Homoscedasticity of error terms
* Uncorrelated error terms
* Independant features

### Creating sample data to compare 

```python
np.random.seed(20)
x = np.arange(20)

x_reshape = x.reshape(-1,1)

y_linear = [x*2 + np.random.rand(1)*4 for x in range(20)]
y_nonlinear = [x**3 + np.random.rand(1)*10 for x in range(20)]
```

### 1. Linear relation between features and the dependant variable

```python
linear = LinearRegression()
linear.fit(x_reshape, y_linear)
```

```python
nonlinear = LinearRegression()
nonlinear.fit(x_reshape, y_nonlinear)
```

#### Understanding the linearity of variables

```python
plt.scatter(x_reshape , y_linear, color = 'red')
plt.plot(x_reshape, linear.predict(x_reshape), color ='blue')
```

```python
plt.scatter(x_reshape , y_nonlinear, color = 'red')
plt.plot(x_reshape, nonlinear.predict(x_reshape), color ='blue')
```

Above is easy to understand because of low dimentionality. We can look at the normality in residuals plot for high dimentional data 

### Residual Plots

```python
fig, axes = plt.subplots(1, 2, sharex=False, sharey=False)
fig.suptitle('[Residual Plots]')
fig.set_size_inches(12,5)

axes[0].plot(linear.predict(x_reshape), y_linear-linear.predict(x_reshape), 'bo')
axes[0].axhline(y=0, color='k')
axes[0].grid()
axes[0].set_title('Linear')
axes[0].set_xlabel('predicted values')
axes[0].set_ylabel('residuals')

axes[1].plot(nonlinear.predict(x_reshape), y_nonlinear-nonlinear.predict(x_reshape), 'ro')
axes[1].axhline(y=0, color='k')
axes[1].grid()
axes[1].set_title('Non-Linear')
axes[1].set_xlabel('predicted values')
```

### Checking the distribution of the residuals

```python
residual_linear = y_linear - linear.predict(x_reshape)
residual_nonlinear = y_nonlinear - nonlinear.predict(x_reshape)
```

```python
# Ploting tbe above residuals to see the distribution

import seaborn as sns

sns.distplot(residual_linear)
plt.title('Linear')
```

```python
sns.distplot(residual_nonlinear)
plt.title('Linear')
```

As we see above, there is a skew in the nonlinear data while the linear data is a normal distribution
