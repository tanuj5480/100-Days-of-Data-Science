```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

%matplotlib inline
```

## Linear Regression Assumptions

* Linear relation between features and the dependant variable
* Homoscedasticity of error terms
* Uncorrelated error terms
* Independant features

### Creating sample data 

```python
np.random.seed(20)
x = np.arange(20)

x_reshape = x.reshape(-1,1)

y_linear = [x*2 + np.random.rand(1)*4 for x in range(20)]
y_nonlinear = [x**3 + np.random.rand(1)*10 for x in range(20)]
```

## 1. Linear relation between features and the dependant variable

```python
linear = LinearRegression()
linear.fit(x_reshape, y_linear)
```

```python
nonlinear = LinearRegression()
nonlinear.fit(x_reshape, y_nonlinear)
```

#### Understanding the linearity of variables

```python
plt.scatter(x_reshape , y_linear, color = 'red')
plt.plot(x_reshape, linear.predict(x_reshape), color ='blue')
```

```python
plt.scatter(x_reshape , y_nonlinear, color = 'red')
plt.plot(x_reshape, nonlinear.predict(x_reshape), color ='blue')
```

Above is easy to understand because of low dimensionality. We can look at the normality in residuals plot for high dimensional data 

### Residual Plots

```python
fig, axes = plt.subplots(1, 2, sharex=False, sharey=False)
fig.suptitle('[Residual Plots]')
fig.set_size_inches(12,5)

axes[0].plot(linear.predict(x_reshape), y_linear-linear.predict(x_reshape), 'bo')
axes[0].axhline(y=0, color='k')
axes[0].grid()
axes[0].set_title('Linear')
axes[0].set_xlabel('predicted values')
axes[0].set_ylabel('residuals')

axes[1].plot(nonlinear.predict(x_reshape), y_nonlinear-nonlinear.predict(x_reshape), 'ro')
axes[1].axhline(y=0, color='k')
axes[1].grid()
axes[1].set_title('Non-Linear')
axes[1].set_xlabel('predicted values')
```

### Checking the distribution of the residuals

```python
residual_linear = y_linear - linear.predict(x_reshape)
residual_nonlinear = y_nonlinear - nonlinear.predict(x_reshape)
```

```python
np.mean(residual_linear)
```

```python
# Ploting tbe above residuals to see the distribution

import seaborn as sns

sns.distplot(residual_linear)
plt.title('Linear')
```

```python
sns.distplot(residual_nonlinear)
plt.title('Linear')
```

As we see above, there is a skew in the nonlinear data while the linear data is a normal distribution 

### Solution for non-linear data
* Feature transformation
* Non-linear algorithms 

## 2. Homoscedasticity of error terms

```python
# generate dummy data

y_homo = [x*2 + np.random.rand(1) for x in range(20)]  ## homoscedastic error
y_hetero = [x*2 + np.random.rand(1)*2*x for x in range(20)]  ## heteroscedastic error
```

```python
# fit model

linear_homo = LinearRegression()
linear_homo.fit(x_reshape, y_homo)
```

```python
linear_hetero = LinearRegression()
linear_hetero.fit(x_reshape, y_hetero)
```

```python
plt.scatter(x_reshape , y_homo, color = 'red')
plt.plot(x_reshape, linear_homo.predict(x_reshape), color ='blue')
```

```python
plt.scatter(x_reshape , y_hetero, color = 'red')
plt.plot(x_reshape, linear_hetero.predict(x_reshape), color ='blue')
```

```python
# Residual Plots

fig, axes = plt.subplots(1, 2, sharex=False, sharey=False)
fig.suptitle('[Residual Plots]')
fig.set_size_inches(12,5)

axes[0].plot(linear_homo.predict(x_reshape), y_homo-linear_homo.predict(x_reshape), 'bo')
axes[0].axhline(y=0, color='k')
axes[0].grid()
axes[0].set_title('Linear')
axes[0].set_xlabel('predicted values')
axes[0].set_ylabel('residuals')

axes[1].plot(linear_hetero.predict(x_reshape), y_hetero-linear_hetero.predict(x_reshape), 'ro')
axes[1].axhline(y=0, color='k')
axes[1].grid()
axes[1].set_title('Non-Linear')
axes[1].set_xlabel('predicted values')
```

Graph on the right reflects Heteroscedasticity, as the error increase with increasing predicted value

### Solution for Heteroscedasticity

* Log transforming the target values

## 3. Uncorrelated error terms

```python
x = np.arange(20)
```

```python
np.random.seed(20)

y_uncorr = [2*x + np.random.rand(1) for x in range(20)]
y_corr = [np.sin(x) for x in range(20)]
```

```python
linear_uncorr = LinearRegression()
linear_uncorr.fit(x_reshape, y_uncorr)
```

```python
linear_corr = LinearRegression()
linear_corr.fit(x_reshape, y_corr)
```

```python
# Residual Plots

fig, axes = plt.subplots(1, 2, sharex=False, sharey=False)
fig.suptitle('[Residual Plots]')
fig.set_size_inches(12,5)

axes[0].plot(linear_uncorr.predict(x_reshape), y_uncorr-linear_homo.predict(x_reshape), 'bo')
axes[0].axhline(y=0, color='k')
axes[0].grid()
axes[0].set_title('Linear')
axes[0].set_xlabel('predicted values')
axes[0].set_ylabel('residuals')

axes[1].plot(linear_corr.predict(x_reshape), y_corr-linear_corr.predict(x_reshape), 'ro')
axes[1].axhline(y=0, color='k')
axes[1].grid()
axes[1].set_title('Non-Linear')
axes[1].set_xlabel('predicted values')
plt.show()
```

### Solution for correlated error terms

* Use time series modeling instead of linear regressionsss

## 4. Independant Features

```python
# generating dummy data

np.random.seed(39)
x1 = np.arange(20) * 2
x2 = np.random.randint(low=0, high=50, size=20)
x_idp = np.vstack((x1,x2))
y = np.add( np.sum(x_idp, axis=0), np.random.randn(20)*5 )  ## y = x1 + x2 + noise
```

```python
# creating linearly dependant feature

import pandas as pd
dp_df = pd.DataFrame([x1,x2,(x1+x2)]).T
```

```python
dp_df.columns = ['x1', 'x2', 'x1+x2']
```

```python
import seaborn as sns

Var_Corr = dp_df.corr()
```

```python
sns.heatmap(Var_Corr, xticklabels=Var_Corr.columns, yticklabels=Var_Corr.columns, annot=True)
```

Problem with dependant features is the difficulty to interpret 

### Solution
* The dot product of any two features is 0 for truly independant variables 
* Larger the magnitude of dot product, higher the correlation
